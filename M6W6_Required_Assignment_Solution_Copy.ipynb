## Task 1: Variables and Data Types
    
#### Write a Python program that:
a) Declares variables for storing a user’s name, age, and height.<br>
b) Prints a greeting message using the name.<br>
c) Calculates the user’s age 10 years from now.<br>
d) Displays height in meters by converting height from centimetres.
##### <i>Explanation:</i>
- declare variables for name (string), age (integer), and height (integer in cm).
- print a greeting using the name variable.
- calculate age in 10 years using simple arithmetic.
- convert height from centimeters to meters (divide by 100).

```python
# Task 1 Solution

# a) Declare variables
name = "Tan Ah Huat"     # string
age = 25                 # integer
height_cm = 171          # integer (centimeters)

# b) Print greeting
print(f"Hello, {name}! Welcome to Python programming.")

# c) Age after 10 years
age_in10 = age + 10
print(f"In 10 years, you will be {age_in10} years old.")

# d) Convert height to meters
height_m = height_cm / 100
print(f"Your height in meters is {height_m} meter.")
```

## Task 2: Conditional Statements


#### Write a Python program that:
a) Prompts the user to enter a number.<br>
b) Checks if the number is positive, negative, or zero.<br>
c) Prints an appropriate message based on the result.<br>

##### <i>Explanation:</i>
- use `input()` to take a number from the user and convert it to integer.
- Then check conditions:
  - If number > 0 → Positive
  - If number < 0 → Negative
  - Else → Zero

```python
# Task 2 Solution

# a) Prompt user input
num = int(input("Enter a number: "))

# b) & c) Check if positive, negative, or zero
if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
```

### Extra Learning: <br>
Improved method to checks for valid number input

```python
print("Number Checker (Positive/Negative/Zero)")
print("---------------------------------------")

while True:
    try:
        # Prompt user input
        num = int(input("Enter a number: "))

        # Check if positive, negative, or zero
        if num > 0:
            print("The number is positive.")
        elif num < 0:
            print("The number is negative.")
        else:
            print("The number is zero.")
        break

    except ValueError:
        print("Please enter a valid whole number. Try again.\n")
```

## Task 3: Loops and Iteration
#### Write a program that:
a) Prompts the user to input an integer nnn.<br>
b) Uses a for loop to print all even numbers from 1 up to nnn.<br>
c) Ensures only even numbers appear in the output.<br>

##### <i>Explanation:</i>
- ask the user for an integer `nnn`.
- use a `for` loop with `range(1, nnn+1)`.
- print only even numbers using modulus operator `%`.

```python
1
# Task 3 Solution
print("Loops and Iteration")
print("-------------------")

# a) Prompt user input
nnn = int(input("Enter an integer: "))

# b) Loop through range and check even numbers
print(f"Even numbers from 1 to {nnn}:")
for i in range(1, nnn+1):
    if i % 2 == 0:   # divisible by 2 means even
        print(i)
```

### Extra Learning: <br>
Improved method to checks for valid number input

```python
print("Even Numbers Finder Method 2")
print("-------------------")

while True:
    try:
        # Prompt user input
        nnn = int(input("Enter an integer: "))

        # Loop through range and check even numbers
        print(f"Even numbers from 1 to {nnn}:")
        for i in range(1, nnn+1):
            if i % 2 == 0:   # divisible by 2 means even
                print(i)
        break

    except ValueError:
        print("Please enter a whole number only. Try again.\n")
```

## Task 4: Functions

#### Write a program that:
a) Defines a function to calculate the area of a rectangle using length and width as inputs. <br>
b) Prompts the user to input values for length and width. <br>
c) Calls the function to calculate and then displays the area in a formatted message. <br>


##### <i>Explanation:<i>
- Define a function `rectangle_area(length, width)` that returns `length * width`.
- Prompt user for length and width.
- Call the function and display the result.

```python
# Task 4 Solution

print("Rectangle Area Calculator")
print("-------------------------")

# a) Define function
def rectangle_area(length, width):
    return length * width

# b) Prompt user input
length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))

# c) Call function and print result
area = rectangle_area(length, width)
print(f"The area of the rectangle is {area} square units.")
```

### Extra Learning: <br>
Improved method to checks for valid number input

```python
print("Rectangle Area Calculator")
print("-------------------------")

def rectangle_area(length, width):
    return length * width

while True:
    try:
        # Get user input
        length = float(input("Enter the length of the rectangle: "))
        width = float(input("Enter the width of the rectangle: "))

        # Calculate and show result
        result = rectangle_area(length, width)
        print(f"The area of the rectangle is {result} square units.")
        break

    except ValueError:  # Only catch number conversion errors
        print("Please enter numbers only. Try again.\n")
```
